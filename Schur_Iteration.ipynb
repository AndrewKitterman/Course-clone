```julia
# Initialize the random number generator
rng = MersenneTwister(2016);
```

```julia
function pretty_print(A)
    m = size(A,1)
    n = size(A,2)
    for i=1:m
        for j=1:n
            @printf "%12.3e" A[i,j]
        end
        @printf "\n"
    end
end    
```

```julia
# Size of matrix
n = 3
X = rand(rng, n, n)
Λ = diagm(Float32[i for i=1:n])
A = X * Λ / X
```

```julia
srand(rng, 16)
Qk = rand(rng, n, n)

for k=1:40
    Qk = A * Qk
    Qk, Rk = qr(Qk)
    if k%10 == 0 || k==1
        println("\nIteration ",k)
        pretty_print(Qk' * A * Qk)
    end
end
```

```julia
n = 16; srand(rng, 16)
X = rand(rng, n, n); Λ = diagm(Float32[i for i=1:n]); 
A = X * Λ / X
niter = 40
Qk = rand(rng, n, n); ev = zeros(niter,n)
for k=1:niter
    Qk = A * Qk
    Qk, Rk = qr(Qk)
    ev[k,:] = diag(Qk' * A * Qk)
end
```

```julia
using PlotlyJS
traces = PlotlyJS.GenericTrace[]
for i=1:n
    traces = [traces; scatter(y=ev[:,i],name=string("entry ",i))]
end
plot(traces)
```
