```julia
n = 128
println("Size of matrix ", n)

# Initialize the random number generator
using Random
rng = MersenneTwister(2016)
```

```julia
# Initialize the matrix
using LinearAlgebra
Qe, Re = qr(rand(n,n))
Re = UniformScaling(1.) + triu(rand(n,n))
@show cond(Re)
# Qe is orthogonal, Re is upper triangular
for i=1:n
    # Multiply row i by 1/2^i
    # Re[i,:] /= 2.0^i
end

# A = Qe * Re
A = Qe * Re
```

```julia
# Compute QR factorization of A
Q, R = qr(A)

# Fix possible differences in the sign of the diagonal entries of R
for i=1:n
    if R[i,i] * Re[i,i] < 0
        R[i,:] = -R[i,:]
        Q[:,i] = -Q[:,i]
    end
end

# We now expect that Q=Qe and R=Re.
```

```julia
# Calculate the error in Q
e = zeros(n)
for j=1:n
    # Processing column j
    e[j] = norm(Q[:,j] - Qe[:,j])
end
```

```julia
using Plots
plotlyjs()
```
